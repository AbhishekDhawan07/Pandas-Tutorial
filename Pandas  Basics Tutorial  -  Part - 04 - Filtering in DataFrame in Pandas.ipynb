# Filtering - Using Conditionals to Filter Rows and Columns.

```python
import pandas as pd
```

### Creating a sample DataFrame

```python
people = {
    "first" : ["Abhishek", "Virat", "Rohit"],
    "last" : ["Dhawan", "Kohli", "Sharma"],
    "email" : ["abhidhawan09@gmail.com", "virat.kohli@gmail.com", "rohit.sharma@gmail.com"]
}
```

```python
df = pd.DataFrame(people)
print(df)
```

```python
# Checking whose last name is "Singh"
filter_last = (df["last"] == "Dhawan")
filter_last
```

```python
print(df[filter_last])
```

```python
# It is same as above
print(df[ df["last"] == "Kohli"])
```

```python
# 3rd way
print(df.loc[filter_last])
```

```python
print(df.loc[filter_last, "email"])
```

### And `&`, Or `|`, Not `~` logical operators

```python
# And operator
print(df[(df["first"]=="Rohit") & (df["last"]=="Sharma")])
```

```python
# Or Operator
print(df[(df["last"]=="Sharma") | (df["last"]=="Dhawan")])
```

```python
# Not Operator (tilde sign ~)
print(df[~(df["last"]=="Dhawan")])
```
