## DataFrame in Pandas Library.

1) A DataFrame is a two-dimensional, size-mutable, and heterogeneous tabular data structure in the Pandas library.

2) It is similar to an Excel spreadsheet or a SQL table — consisting of Rows and Columns.

```python
# Example of DataFrame.

import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'London', 'Paris']}

df = pd.DataFrame(data)
print(df)
```

## DataFrame Structure.

- A DataFrame has 3 main components:

1) Data(The Actual Values).

2) Rows(Identified by Index).

3) Columns(Identified by Column Labels).

```python
# We can access these parts easily by.

print(df.columns)  # Column labels
print(df.index)    # Row indices
print(df.values)   # Data as a NumPy array
```

## Different Ways for Creating a DataFrame.

## From a Dictionary of Lists.

```python
# Syntax : df = pd.DataFrame({'col1': [v1, v2], 'col2': [v3, v4]})
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'London', 'Paris']
}
df = pd.DataFrame(data)
print(df)
```

## From a List of Dictionaries.

```python
# Syntax: df = pd.DataFrame([{'col1': v1, 'col2': v2}, {'col1': v3, 'col2': v4}])
```

```python
data = [
    {'Name': 'Alice', 'Age': 25},
    {'Name': 'Bob', 'Age': 30}
]
df = pd.DataFrame(data)
print(df)
```

## From a Dictionary of Series.

```python
# Syntax: df = pd.DataFrame({'col1': pd.Series(data1), 'col2': pd.Series(data2)})
```

```python
data = {
    'A': pd.Series([10, 20, 30], index=['a', 'b', 'c']),
    'B': pd.Series([100, 200, 300], index=['a', 'b', 'c'])
}
df = pd.DataFrame(data)
print(df)
```

## From a 2D List(List of Lists).

```python
# Syntax: df = pd.DataFrame([[v1, v2], [v3, v4]], columns=['col1', 'col2'])
```

```python
data = [[1, 'Alice'], [2, 'Bob'], [3, 'Charlie']]
df = pd.DataFrame(data, columns=['ID', 'Name'])
print(df)
```

## From a NumPy Array.

```python
# Syntax: df = pd.DataFrame(numpy_array, columns=['col1', 'col2'])
```

```python
import numpy as np

data = np.array([[10, 20], [30, 40], [50, 60]])
df = pd.DataFrame(data, columns=['X', 'Y'])
print(df)
```

## Features of Pandas DataFrame.

## Heterogeneous Data(Multiple Data Types).

- A DataFrame can store different data types in different columns — integers, floats, strings, booleans, etc.

```python
# Example:

import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000.50, 60000.75, 70000.00],
    'Is_Manager': [False, True, False]
}

df = pd.DataFrame(data)
print(df.dtypes)
```

## Labeled Axes(Rows and Columns).

- Each column and row has a label for easy identification and access.

```python
# Example:

print(df.columns)  # Column labels
print(df.index)    # Row index labels
```

## Size-Mutable(Can Add or Remove Columns/Rows).

- You can dynamically modify a DataFrame by adding or deleting columns or rows.

```python
# Example:

# Add new column
df['Department'] = ['HR', 'IT', 'Finance']

# Remove column
df.drop('Is_Manager', axis=1, inplace=True)

print(df)
```

## Automatic and Custom Indexing.

- Pandas automatically assigns Numeric Indexes, but you can also define custom ones.

```python
# Example:

df = pd.DataFrame(data, index=['E1', 'E2', 'E3'])
print(df)
```

## Data Selection and Filtering.

- You can select, filter, or slice data easily using Labels or Conditions.

```python
# Example:

# Select a column
print(df['Name'])

# Filter rows
filtered = df[df['Age'] > 28]
print(filtered)
```

## Handling Missing Data(NaN Values).

- DataFrames have built-in methods for detecting, removing, or filling missing data.

```python
# Example:

import numpy as np

df2 = pd.DataFrame({
    'A': [1, 2, np.nan],
    'B': [4, np.nan, 6]
})

print(df2.fillna(0))  # Replace NaN with 0.
```

## Data Alignment and Broadcasting.

- Pandas automatically aligns data based on labels when performing operations.

```python
# Example:

df_a = pd.DataFrame({'X': [10, 20, 30]}, index=['a', 'b', 'c'])
df_b = pd.DataFrame({'X': [1, 2, 3]}, index=['b', 'c', 'd'])

print(df_a + df_b)
```

## Powerful Aggregation and Statistical Functions

- You can perform mathematical and statistical operations column-wise or row-wise.

```python
# Example:

df = pd.DataFrame({
    'Sales_Q1': [100, 150, 200],
    'Sales_Q2': [120, 180, 210]
}, index=['A', 'B', 'C'])

print(df.mean())      # Column-wise mean
print(df.sum(axis=1)) # Row-wise sum
```

## Input/Output Operations(I/O).

- DataFrames support multiple File Formats for Reading and Writing.

```python
# Example: 

# Write to CSV
df.to_csv('sales.csv', index=False)

# Read from CSV
new_df = pd.read_csv('sales.csv')
print(new_df)
```

## Integration with NumPy, Matplotlib, and Other Libraries.

- Pandas works well with other Data Science Libraries.

```python
# Example: 

import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar'],
    'Revenue': [1000, 1500, 1300]
})

plt.plot(df['Month'], df['Revenue'])
plt.title("Monthly Revenue")
plt.show()
```

```python
## 10 Examples of Pandas DataFrame.
```

```python
# Example 1: Creating a Simple DataFrame.

import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)
print(df)
```

```python
# Example 2: Adding a New Column.

df['City'] = ['New York', 'London', 'Paris']
print(df)
```

```python
# Example 3: Selecting Specific Columns and Rows.

print(df[['Name', 'City']])     # Select columns
print(df.iloc[0:2])             # Select first two rows
```

```python
# Example 4: Filtering Rows Based on Condition.

filtered_df = df[df['Age'] > 28]
print(filtered_df)
```

```python
# Example 5: Handling Missing Values.

import numpy as np

df2 = pd.DataFrame({'A': [1, 2, np.nan], 'B': [4, np.nan, 6]})
print(df2.fillna(0))
```

```python
# Example 6: Performing Mathematical Operations.

df_math = pd.DataFrame({'X': [10, 20, 30], 'Y': [5, 10, 15]})
df_math['Sum'] = df_math['X'] + df_math['Y']
df_math['Average'] = df_math[['X', 'Y']].mean(axis=1)
print(df_math)
```

```python
# Example 7: Sorting Data.

df_sorted = df.sort_values(by='Age', ascending=False)
print(df_sorted)
```

```python
# Example 8: Grouping and Aggregation.

data = {
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR'],
    'Salary': [40000, 60000, 50000, 65000, 42000]
}
df_group = pd.DataFrame(data)

grouped = df_group.groupby('Department')['Salary'].mean()
print(grouped)


```

```python
# Example 9: Merging Two DataFrames.

df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 3], 'Salary': [50000, 60000, 70000]})

merged_df = pd.merge(df1, df2, on='ID')
print(merged_df)
```

```python
# Example 10: Visualization from DataFrame.

import matplotlib.pyplot as plt

df_plot = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
    'Sales': [300, 400, 350, 500]
})

df_plot.plot(x='Month', y='Sales', kind='bar', title='Monthly Sales', legend=False)
plt.show()
```
